```python
import numpy as np
import pandas as pd

human_pop = pd.read_csv(r'상권_상주인구.csv')
```

```python
human_pop = pd.DataFrame(human_pop)
```

```python
human_population =  human_pop.drop(['Unnamed: 0'], axis=1)
```

```python
human_population
```

```python
apt = pd.read_csv(r'상권_아파트.csv')
```

```python
apart = apt.drop(['Unnamed: 0'], axis=1)
```

```python
apart
```

```python
market_info = pd.merge(human_population,apart,how='outer',on=['상권_코드','기준_년_코드','기준_분기_코드','상권_구분_코드','상권_구분_코드_명'])
market_info
```
