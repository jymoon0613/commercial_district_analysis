## 데이터 수집 및 전처리

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("data_상권/서울시 우리마을가게 상권분석서비스(상권-직장인구)_new.csv",encoding='cp949')
data
```

```python
data.groupby(['기준_년월_코드','기준_분기_코드']).size()
```

```python
data.groupby(['기준_년월_코드','기준_분기_코드']).get_group((2014,4))
```

```python
data=data.drop(data.index[30369:34382])
data
```

```python
data=data.drop(['Unnamed: 0','기준_분기_코드.1'],axis=1,inplace=False)
data
```

```python
data.groupby(['기준_년월_코드','기준_분기_코드']).size()
```

```python
data=data.rename({'기준_년월_코드':'기준_년_코드'},axis='columns')
data
```

```python
data_jy=pd.read_csv("data_상권/merge_df_문지용.csv",encoding='cp949')
data_jy
```

```python
data_jy=data_jy.drop(['Unnamed: 0'],axis=1,inplace=False)
data_jy
```

```python
data_jy=data_jy.drop(['상권_구분_코드'],axis=1,inplace=False)
data_jy
```

```python
data.columns
```

```python
data_jy.columns
```

```python
data_all=pd.merge(data_jy,data,how='left',on=['상권_구분_코드_명', '상권_코드', '상권_코드_명', '기준_년_코드', '기준_분기_코드'])
data_all
```

```python
data_sw=pd.read_csv("data_상권/merged_상주인구_아파트.csv",encoding='cp949')
data_sw
```

```python
data_sw=data_sw.drop(['Unnamed: 0','상권_구분_코드'],axis=1,inplace=False)
data_sw
```

```python
data_sw.columns
```

```python
data_all_3=pd.merge(data_all,data_sw,how='left',on=['상권_구분_코드_명', '상권_코드', '상권_코드_명', '기준_년_코드', '기준_분기_코드'])
data_all_3
```

```python
data_all_3.columns
```

```python
data_all_3.to_csv('data_processed_final.csv')
```
