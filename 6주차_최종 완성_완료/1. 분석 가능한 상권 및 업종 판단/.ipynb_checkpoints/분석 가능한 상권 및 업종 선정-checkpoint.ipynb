# Ⅰ. 분석 가능한 상권 및 업종 선정

```python
# 필요한 라이브러리 불러오기
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

## 1. 데이터 불러오기

```python
# 데이터 경로 설정
path1 = './통합_점포_추정매출_최종.csv'

# 2015-1분기 ~ 2020-2분기까지의 모든 상권 및 모든 서비스업종에 관한 데이터
df_total = pd.read_csv(path1, encoding = 'cp949').drop('Unnamed: 0', axis = 1)

# 데이터 경로 설정
path2 = './결측치_비율계산_최종.csv'

# 2015-1분기 ~ 2020-2분기까지의 모든 상권 및 모든 서비스업종에 관한 데이터에서 결측치 비율을 계산한 데이터
df_null = pd.read_csv(path2, encoding = 'cp949').drop(['Unnamed: 0', '점포수'], axis = 1)
```

```python
# 결과 확인
df_total.head()
```

## 2. 분석 가능한 상권 및 업종 선정

```python
# 점포+매출 통합 데이터에서 모든 column별 결측치의 수 탐색
for col in df_total.columns:
    n_of_null = df_total[col].isnull().sum()
    print('{} : {}'.format(col, n_of_null))
```

- 매출데이터가 하나라도 누락된 경우, 모든 매출 데이터가 누락됨
- 결측치가 많은 열을 제거하기에는 잃게 되는 정보가 매우 많음  
- 따라서 행을 기준으로 결측치를 처리해야 함

### (1) 점포+매출 데이터 두 가지 모두 누락된 경우를 제거

```python
# 점포+매출 데이터가 모두 누락된 경우에는 데이터 상에 행 자체가 존재하지 않기 때문에 결측치 비율로는 찾을 수 없음
# 두 데이터가 모두 누락된 경우, 해당 상권과 업종을 추출했을 때 행의 길이가 22가 되지 않음
# 2015년 1분기 ~ 2020년 2분기 = 총 22개 행
# 점포 데이터가 누락되었다는 것은 과거 폐업률 수치를 확인할 수 없다는 것을 의미함
# 과거 폐업률은 모델 학습의 기본이자 핵심이 되는 정보이기 때문에 완전하지 않은 경우 해당 데이터를 삭제하는 것이 적합함

# 제거할 대상의 index를 저장하기 위한 리스트 생성
delete_list_for_total = []
delete_list_for_null = []

market_code_list = list(df_null['상권코드'])
service_code_list = list(df_null['업종코드'])

for m,s in zip(market_code_list, service_code_list):
    select_market = df_total[df_total['상권_코드'] == m]
    select_service = select_market[select_market['서비스_업종_코드'] == s]
    if len(select_service) != 22:
        delete_list_for_total.append(list(select_service.index))
        select_market_2 = df_null[df_null['상권코드'] == m]
        select_service_2 = select_market_2[select_market_2['업종코드'] == s]
        delete_list_for_null.append(list(select_service_2.index))
    else:
        continue
```

```python
# flattening list

delete_list_for_total_index = []
delete_list_for_null_index = []

for index_list in delete_list_for_total:
    for ind in index_list:
        delete_list_for_total_index.append(ind)
        
for index_list in delete_list_for_null:
    for ind in index_list:
        delete_list_for_null_index.append(ind)
```

```python
# 해당되는 데이터 제거
    
df_total_new = df_total.drop(delete_list_for_total_index).reset_index(drop = True)
    
df_null_new = df_null.drop(delete_list_for_null_index).reset_index(drop = True)
```

```python
# 제거 후 결과 확인 
df_total_new
```

```python
# 제거 후 결과 확인
df_null_new
```

### (2) 결측치 비율에 따른 분석 대상 선정

```python
# 결측비율이 10% 이상인 상권과 업종 추출
target_null_ratio = 0.10
null_10 = df_null_new[df_null_new['결측비율'] >= target_null_ratio]
```

```python
# 결측비율이 10% 이상인 상권과 업종 중, 최대 결측비율을 갖는 경우 추출

max_null = null_10['결측비율'].max()

# 결측치 비율이 최대인 상권 추출
max_null_market = null_10[null_10['결측비율'] == max_null]

# 결과출력
max_null_market
```

```python
# 점포+매출 통합 데이터에서 결측비율이 85%이상인 경우 실제로 어떻게 나타나는지 탐색

# 상권코드 = 1001274, 업종코드 = CS300008인 데이터 추출
case1 = df_total_new[df_total_new['상권_코드'] == 1001274]
case1 = case1[case1['서비스_업종_코드'] == 'CS300008']

# 결과 출력
case1
```

- 결측비율이 85% 이상인 경우, 모델의 input으로 사용되는 모든 column이 결측되어 있음. (연도, 분기별로 모두 결측) 
- 이 경우, 폐업률 예측 이전에 모든 input 데이터부터 추정해야 하기 때문에 폐업률 예측의 의미가 없음. (예측의 예측이 되어버린다.)

```python
# 결측비율이 10% 이상인 상권과 업종 중, 최소 결측비율을 갖는 경우 추출

min_null = null_10['결측비율'].min()

# 결측치 비율이 최소인 상권 추출
min_null_market = null_10[null_10['결측비율'] == min_null]

# 결과출력
min_null_market
```

- 결측치가 없는 상권 및 업종만을 분석 대상으로 선택

### (3) 분석 대상 상권 및 업종만 추출

```python
# 결측치가 없는 상권과 업종 추출
target_null_ratio = 0
null_0 = df_null_new[df_null_new['결측비율'] == target_null_ratio]

select_index = []

market_code_list = list(null_0['상권코드'])
service_code_list = list(null_0['업종코드'])

for m,s in zip(market_code_list, service_code_list):
    select_market = df_total_new[df_total_new['상권_코드'] == m]
    select_service = select_market[select_market['서비스_업종_코드'] == s]
    select_index.append(list(select_service.index))
```

```python
# flattening list
select_index_list = []

for index_list in select_index:
    for ind in index_list:
        select_index_list.append(ind)
```

```python
# 전체 데이터에서 분석 대상 최종 선택

df_total_final = df_total_new.loc[select_index_list,:].reset_index(drop = True)
```

```python
# 결과확인

df_total_final
```

- 최종적으로 총 59만개 정도의 상권과 업종이 선택됨

```python
# 결측치 확인

df_total_final.isnull().sum().sum()
```
