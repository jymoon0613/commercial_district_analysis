```python
# 선택할 데이터의 index를 저장하기 위한 list 정의
fluid_select_index = []
facil_select_index = []
work_select_index = []
live_select_index = []

# 유동인구
for m in market_code_list:
    select_market = df_fluid_new[df_fluid_new['상권_코드'] == m]
    fluid_select_index.append(list(select_market.index))

# 집객시설
for m in market_code_list:
    select_market = df_facil_new[df_facil_new['상권_코드'] == m]
    facil_select_index.append(list(select_market.index))
    
# 직장인구
for m in market_code_list:
    select_market = df_work_new[df_work_new['상권_코드'] == m]
    work_select_index.append(list(select_market.index))
    
# 상주인구
for m in market_code_list:
    select_market = df_live_new[df_live_new['상권_코드'] == m]
    live_select_index.append(list(select_market.index))
```

```python
# flattening list

fluid_select_index_total = []
facil_select_index_total = []
work_select_index_total = []
live_select_index_total = []
# 유동인구
for index_list in fluid_select_index:
    for ind in index_list:
        fluid_select_index_total.append(ind)

# 집객시설
for index_list in facil_select_index:
    for ind in index_list:
        facil_select_index_total.append(ind)
        
# 직장인구
for index_list in work_select_index:
    for ind in index_list:
        work_select_index_total.append(ind)

# 상주인구
for index_list in live_select_index:
    for ind in index_list:
        live_select_index_total.append(ind)
```
