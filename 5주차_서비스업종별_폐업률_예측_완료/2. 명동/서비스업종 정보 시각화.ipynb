## 2. '명동거리' - '일반의류' 업종 데이터 시각화
### (1) 점포 현황

```python
# 업종 선택

cloth = df_m['서비스_업종_코드_명'] == '일반의류'

df_m_cloth = df_m[cloth].reset_index(drop = True)

df_m_cloth.head()
```

```python
# 일반 점포수 추가 (전체 점포 수 - 프랜차이즈 점포 수)

df_m_cloth['일반_점포_수'] = 0

for i in range(len(df_m_cloth)):
    n = df_m_cloth.loc[i, '점포_수']
    f = df_m_cloth.loc[i, '프랜차이즈_점포_수']
    df_m_cloth.loc[i,'일반_점포_수'] = n-f
```

```python
# 시계열 Lineplot을 위한 column 추가

# 연도, 분기 추출
year = list(df_m_cloth['기준_년_코드'])
month = list(df_m_cloth['기준_분기_코드'])

# (연도, 분기) list 생성
year_month = [i for i in zip(year,month)]
year_month

# '연도_분기' list 생성
# 1분기 = 3월 1일, 2분기 = 6월 1일, 3분기 = 9월 1일, 4분기 = 12월 1일
y_m = []
for i in range(len(year_month)):
    month = ''
    if int(year_month[i][1]) == 1:
        month = 3
    elif int(year_month[i][1]) == 2:
        month = 6
    elif int(year_month[i][1]) == 3:
        month = 9
    else :
        month = 12
    a = str(int(year_month[i][0])) + '-' + str(month) + '-1'
    y_m.append(a)
    
# 새로운 column 추가
df_date = df_m_cloth.copy()
df_date['Date'] = y_m
df_date['Date'] = df_date['Date'].astype('datetime64[ns]')
df_date = df_date.set_index('Date')
df_date
```

```python
df_date.columns
```

```python
# 일반의류 점포 현황

df_2020 = df_date[df_date['기준_년_코드'] == 2020]
df_2020_2 = df_2020[df_2020['기준_분기_코드'] == 2]
total = df_2020_2['점포_수']
gen = df_2020_2['일반_점포_수']
fran = df_2020_2['프랜차이즈_점포_수']

fig, ax = plt.subplots(figsize=(8,5))
sns.barplot(['총 점포 수', '일반 점포', '프랜차이즈 점포'], [total,gen,fran], ax = ax)
ax.set_title('"일반 의류" 점포 현황')

for p in ax.patches:
    left, bottom, width, height = p.get_bbox().bounds
    ax.annotate("총 {:,d} 개".format(int(height)), (left+width/2, height*1.01), ha='center')
```

```python
# 남녀 매출 비율 (2020-2)

import numpy as np

df_2020 = df_date[df_date['기준_년_코드'] == 2020]
df_2020_2 = df_2020[df_2020['기준_분기_코드'] == 2]
dff = df_2020_2[['남성_매출_비율', '여성_매출_비율']]
aa = pd.DataFrame()

for col in dff.columns:
    if int(dff[col]) == 0:
        continue
    else:  
        n = int(dff[col])
        word = col.split('_')[0]
        w_list = []
        w_list.append(word)
        ww = w_list*n
        aa = aa.append(ww)

fig, ax = plt.subplots(figsize=(5,5))
df = aa.loc[:,0].value_counts()


df.plot.pie(title = '남녀 매출 비율', autopct='%.1f%%', label = '')
```
