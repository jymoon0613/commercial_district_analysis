## 3. 추출한 데이터 통합

```python
# 공통되는 column 찾기

fluid = set(fluid_m.columns)
facil = set(facil_m.columns)
work = set(work_m.columns)
live = set(live_m.columns)

r1 = set.intersection(fluid, facil)
r2 = set.intersection(r1, work)
r3 = set.intersection(r2, live)

print("{}\n{}\n{}\n".format(r1, r2, r3))
```

```python
# 데이터 merge

t1 = pd.merge(fluid_m, facil_m, how = 'left', on = list(r1))
t2 = pd.merge(t1, work_m, how = 'left', on = list(r2))
total = pd.merge(t2, live_m, how = 'left', on = list(r3))
```

```python
# 결과 확인

total
```

```python
# 결측치 확인

total.isnull().sum().sum()
```

```python
# 불필요한 column 제거
useless = [' 상권_구분_코드', ' 상권_구분_코드_명', '상권_구분_코드_명', '상권_구분_코드_x', '상권_구분_코드_명_x', '상권_구분_코드_명_y', 
           '상권_구분_코드_y']

total_new = total.drop(useless, axis = 1)
```

## 4. 데이터 정규화

```python
# Min-Max 정규화 진행

from sklearn import preprocessing

for i in range(4,72):
    if total_new.columns[i] == '폐업_률':
        continue
    else:
        x = total_new[total_new.columns[i]].values # Return numpy array
        min_max_scaler = preprocessing.MinMaxScaler()
        x_scaled = min_max_scaler.fit_transform(pd.DataFrame(x))
        # 적용
        total_new[total_new.columns[i]] = pd.DataFrame(x_scaled)
```
