## 2. 누락 상권 조사
### (1) 유동인구

```python
# 데이터셋 연도별, 분기별 분할
start_year = 2015 # 시작 연도
end_year = 2020 # 끝 연도
start_quarter = 1 # 시작 분기
end_quarter = 4 # 끝 분기

for i in range(start_year, end_year + 1):
    df = df1[df1['기준_년_코드'] == i]
    for j in range(start_quarter, end_quarter + 1):
        globals()['df1_{}_{}'.format(i, j)] = df[df['기준_분기_코드'] == j]
        if i == start_year and j == end_quarter - 2: # 2020년은 2분기까지 밖에 없으므로 조기 종료
            break
```

```python
# 연도, 분기별 상권 개수 출력

for i in range(start_year, end_year + 1):
    for j in range(start_quarter, end_quarter + 1):
        print(len(globals()['df1_{}_{}'.format(i, j)]))
        if i == start_year and j == end_quarter - 2: # 2020년은 2분기까지 밖에 없으므로 조기 종료
            break
```

```python
# 어떤 상권이 빠졌는지 조사

for i in range(start_year, end_year + 1):
    for j in range(start_quarter, end_quarter + 1):
        globals()['s1_{}_{}'.format(i, j)] = set(globals()['df1_{}_{}'.format(i, j)]['상권_코드'])
        if i == start_year and j == end_quarter - 2: # 2020년은 2분기까지 밖에 없으므로 조기 종료
            break
            
for i in range(start_year, end_year + 1):
    for j in range(start_quarter, end_quarter + 1):
        globals()['d1_{}_{}'.format(i, j)] = globals()['s1_{}_{}'.format(i, j)]
        if i == start_year and j == end_quarter - 2: # 2020년은 2분기까지 밖에 없으므로 조기 종료
            break
```

```python
# 연도별, 분기별 빠진 상권의 상권 코드 출력

for i in range(start_year, end_year + 1):
    for j in range(start_quarter, end_quarter + 1):
        print("{}_{} : {}".format(i, j, globals()['d1_{}_{}'.format(i, j)]))
        if i == start_year and j == end_quarter - 2: # 2020년은 2분기까지 밖에 없으므로 조기 종료
            break
```

```python
# 누락된 상권의 상권명 출력

market_code_list = list(df['상권_코드'].unique())
market_name_list = list(df['상권_코드_명'].unique())

# 누락된 상권 코드 저장
total_code_list = [] 
for i in range(start_year, end_year + 1):
    for j in range(start_quarter, end_quarter + 1):
        total_code_list.append(list(globals()['d1_{}_{}'.format(i, j)]))
        if i == start_year and j == end_quarter - 2: # 2020년은 2분기까지 밖에 없으므로 조기 종료
            break
            
# 상권 코드에 대응하는 상권명 저장
total_code_list = []
for code_list in total_code_list:
    for code in code_list:
        name_list = []
        ind = marekt_code_list.index(code)
        name = market_name_list[ind]
        name_list.append(name)
    total_code_list.append(name_list)
    

# 결과 출력
n = 0
for i in range(start_year, end_year + 1):
    for j in range(start_quarter, end_quarter + 1):
        print("{}_{} : {}".format(i, j, total_code_list[n]))
        if i == start_year and j == end_quarter - 2: # 2020년은 2분기까지 밖에 없으므로 조기 종료
            break
```

- 유동인구는 KT LTE 시그널 데이터를 이용하여 기지국 당 특정 시점의 생활인구를 구하고 이를 전체 인구 기준으로 변환하여 기지국 범위 내의 구역에 배분하는 방식  
- 해당 상권에 해당하는 기지국을 찾고 직접 값을 구하는 방식은 매우 복잡해 보이고, KT에서도 기지국에 기록된 생활인구를 어떻게 각 지역에 배분하는지는 공개하지 않음  
- 누락된 상권과 인접한 상권의 유동인구로 대체하는 방식이 더 좋아 보임  
- 하지만 서울 우리마을가게 상권 분석 서비스에서는 누락 상권의 유동인구를 0으로 취급함
