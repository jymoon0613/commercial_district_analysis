```python
# 상권 번호를 상권 이름으로 변환하기
# 상권 번호와 상권 이름을 매칭 후, 상권 이름을 리턴한다.
# 우선 빠진 상권 번호를 구한다.


def 해당_년도_분기에_빠진_상권(ba,data_set):
    temp_list =[]
    for i in range (1000001,1001497):
        if (data_set['상권_코드']==i).any():
            pass
        else:
            temp_list.append(i)
    
    
    name_list =[]
    # 누락된 상권 번호의 개수만큼 이름과 매칭한다
    for i in range(0,len(temp_list)):
        name_list.append((ba[(ba['상권_코드']==int(temp_list[i]))]['상권_코드_명']).tolist()[0])
        
    
    return (name_list)
```

```python
for i in range (2015,2020): # 년도
    for j in range(1,5): # 분기
        print(str(i)+'년도 '+str(j)+
              '분기 아파트 데이터의 결측 상권 : \n'+
              str(해당_년도_분기에_빠진_상권(ba,apt[(apt['기준_분기_코드']==j) & (apt['기준_년_코드']==i)]))+
              '\n총 '+
              str(len(해당_년도_분기에_빠진_상권(ba,apt[(apt['기준_분기_코드']==j) & (apt['기준_년_코드']==i)])))+
              ' 개\n')
        
for i in range (2020,2021): # 년도
    for j in range(1,3): # 분기
        print(str(i)+'년도 '+str(j)+'분기 아파트 데이터의 결측 상권 : \n'+
              str(해당_년도_분기에_빠진_상권(ba,apt[(apt['기준_분기_코드']==j) & (apt['기준_년_코드']==i)]))+
              '\n총 '+
              str(len(해당_년도_분기에_빠진_상권(ba,apt[(apt['기준_분기_코드']==j) & (apt['기준_년_코드']==i)])))+
              ' 개\n')
```
